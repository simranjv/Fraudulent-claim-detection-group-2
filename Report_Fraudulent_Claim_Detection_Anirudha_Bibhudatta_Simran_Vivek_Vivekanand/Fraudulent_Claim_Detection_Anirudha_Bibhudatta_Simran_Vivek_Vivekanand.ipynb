## **2. Data Cleaning** <font color = red>[10 marks]</font>

### **2.1 Handle null values** <font color = red>[2 marks]</font>

#### **2.1.1** Examine the columns to determine if any value or column needs to be treated <font color="red">[1 Mark]</font>

```python
# Check the number of missing values in each column
print("Missing values per column:")
print(df.isnull().sum())

print("\nColumns with '?' values (used as null indicator):")
for col in df.columns:
    q_count = (df[col] == '?').sum()
    if q_count > 0:
        print(f"  {col}: {q_count} '?' values")
```

#### **2.1.2** Handle rows containing null values <font color="red">[1 Mark]</font>

```python
# Handle the rows containing null values
# Step 1: First remove completely empty columns (all NaN) so they don't interfere with row-level dropna
all_nan_cols = df.columns[df.isnull().all()].tolist()
if all_nan_cols:
    df.drop(columns=all_nan_cols, inplace=True)
    print(f"Dropped completely empty columns: {all_nan_cols}")

# Step 2: Replace '?' with NaN (it is used as a missing value indicator in this dataset)
df.replace('?', np.nan, inplace=True)

# Step 3: For categorical columns with NaN, fill with mode
cat_cols_with_nan = [col for col in df.select_dtypes(include=['object']).columns
                     if df[col].isnull().any()]
for col in cat_cols_with_nan:
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)
    print(f"Filled NaN in '{col}' with mode: '{mode_val}'")

# Step 4: Drop any remaining rows with null values
rows_before = len(df)
df.dropna(inplace=True)
print(f"\nDropped {rows_before - len(df)} rows with remaining null values")
print(f"Dataset shape after handling null values: {df.shape}")
```

### **2.2 Identify and handle redundant values and columns** <font color = red>[5 marks]</font>

#### **2.2.1** Examine the columns to determine if any value or column needs to be treated <font color="red">[2 Mark]</font>

```python
# Write code to display all the columns with their unique values and counts and check for redundant values
for col in df.columns:
    print(f"\nColumn: {col} | Unique values: {df[col].nunique()} | Dtype: {df[col].dtype}")
    print(df[col].value_counts(dropna=False).head(5).to_string())
    print("=" * 60)
```

#### **2.2.2** Identify and drop any columns that are completely empty <font color="red">[1 Mark]</font>

```python
# Identify and drop any columns that are completely empty
empty_cols = df.columns[df.isnull().all()].tolist()
print(f"Completely empty columns: {empty_cols}")

df.drop(columns=empty_cols, inplace=True)
print(f"Dataset shape after dropping empty columns: {df.shape}")
```

#### **2.2.3** Identify and drop rows where features have illogical or invalid values, such as negative values for features that should only have positive values <font color="red">[1 Mark]</font>

```python
# Identify and drop rows where features have illogical or invalid values,
# such as negative values for features that should only have positive values

print("Checking for negative values in numeric columns:")
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for col in num_cols:
    neg_count = (df[col] < 0).sum()
    if neg_count > 0:
        print(f"  {col}: {neg_count} negative values")

# Drop rows with negative values in columns that must be non-negative
cols_must_positive = ['age', 'months_as_customer', 'policy_deductable',
                      'policy_annual_premium', 'total_claim_amount',
                      'injury_claim', 'property_claim', 'vehicle_claim',
                      'number_of_vehicles_involved', 'bodily_injuries', 'witnesses']

rows_before = len(df)
for col in cols_must_positive:
    if col in df.columns:
        df = df[df[col] >= 0]

df.reset_index(drop=True, inplace=True)
print(f"\nDropped {rows_before - len(df)} rows with illogical/invalid values")
print(f"Dataset shape: {df.shape}")
```

#### **2.2.4** Identify and remove columns where a large proportion of the values are unique or near-unique, as these columns are likely to be identifiers or have very limited predictive power <font color="red">[1 Mark]</font>

```python
# Identify and remove columns that are likely to be identifiers or have very limited predictive power
print("Checking uniqueness ratio of each column:")
for col in df.columns:
    unique_ratio = df[col].nunique() / len(df)
    print(f"  {col}: {df[col].nunique()} unique values ({unique_ratio:.2%})")

# Drop identifier columns and high-cardinality columns with very limited predictive value
# - policy_number: unique identifier for each policy
# - insured_zip: high-cardinality geo code (near-unique)
# - incident_location: near-unique street address
# - auto_model: high cardinality, limited predictive power
cols_to_drop = ['policy_number', 'insured_zip', 'incident_location', 'auto_model']
df.drop(columns=cols_to_drop, inplace=True)
print(f"\nDropped columns: {cols_to_drop}")
print(f"Dataset shape after dropping identifier columns: {df.shape}")
```

### **4.4 Perform bivariate analysis** <font color="red">[10 Marks]</font>

#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features
def target_likelihood_analysis(df_features, df_target, target_label='Y', title_suffix=''):
    cat_cols = df_features.select_dtypes(include=['object']).columns.tolist()
    # Exclude datetime columns
    date_cols = df_features.select_dtypes(include=['datetime64']).columns.tolist()
    cat_cols = [c for c in cat_cols if c not in date_cols]

    for col in cat_cols:
        combined = pd.concat([df_features[col].reset_index(drop=True),
                               df_target.reset_index(drop=True)], axis=1)
        combined.columns = [col, 'fraud_reported']

        likelihood = combined.groupby(col)['fraud_reported'].apply(
            lambda x: (x == target_label).sum() / len(x)
        ).reset_index()
        likelihood.columns = [col, 'fraud_likelihood']
        likelihood = likelihood.sort_values('fraud_likelihood', ascending=False)

        overall_rate = (df_target == target_label).mean()

        plt.figure(figsize=(10, 4))
        sns.barplot(x=col, y='fraud_likelihood', data=likelihood)
        plt.axhline(y=overall_rate, color='red', linestyle='--', label=f'Overall avg ({overall_rate:.2%})')
        plt.title(f'Fraud Likelihood by {col} {title_suffix}')
        plt.xticks(rotation=45, ha='right')
        plt.ylabel('P(Fraud=Y)')
        plt.legend()
        plt.tight_layout()
        plt.show()
        print(likelihood.to_string(index=False))
        print("=" * 60)

target_likelihood_analysis(X_train, y_train, title_suffix='(Training Data)')
```

### **7.3 Find the Optimal Cutoff** <font color = red>[12 marks]</font>
Find the optimal cutoff to improve model performance by evaluating various cutoff values and their impact on relevant metrics.

#### **7.3.1** Plot ROC Curve  to visualise the trade-off between true positive rate and false positive rate across different classification thresholds <font color="red">[2 Marks]</font>

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, auc

# Define ROC function
def draw_roc(actual, probs, title='ROC Curve'):
    fpr, tpr, thresholds = roc_curve(actual, probs, pos_label=1)
    auc_score = auc(fpr, tpr)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc_score:.4f})', color='darkorange', lw=2)
    plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
    plt.xlabel('False Positive Rate (1 - Specificity)')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title(title)
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()
    print(f"AUC Score: {auc_score:.4f}")
    return fpr, tpr, thresholds
```

```python
# Call the ROC function
fpr, tpr, thresholds = draw_roc(y_train_pred_df['Actual'], y_train_pred_df['Predicted_Prob'],
                                 title='ROC Curve - Logistic Regression (Training Data)')
```

#### **7.3.2** Predict on training data at various probability cutoffs <font color="red">[1 Mark]</font>

```python
# Create columns with different probability cutoffs to explore the impact of cutoff on model performance
for cutoff in [0.3, 0.4, 0.5, 0.6, 0.7]:
    y_train_pred_df[f'Predicted_{cutoff}'] = (y_train_pred_df['Predicted_Prob'] >= cutoff).astype(int)

y_train_pred_df.head()
```

#### **7.3.3** Plot accuracy, sensitivity, specificity at different values of probability cutoffs <font color="red">[2 Marks]</font>

```python
# Create a DataFrame to see the values of accuracy, sensitivity, and specificity at different values of probability cutoffs
cutoff_values = np.arange(0.01, 1.0, 0.01)
results_list = []

for cutoff in cutoff_values:
    predicted = (y_train_pred_df['Predicted_Prob'] >= cutoff).astype(int)
    actual = y_train_pred_df['Actual']
    tn, fp, fn, tp = metrics.confusion_matrix(actual, predicted, labels=[0, 1]).ravel()
    acc  = (tp + tn) / (tp + tn + fp + fn)
    sens = tp / (tp + fn) if (tp + fn) > 0 else 0
    spec = tn / (tn + fp) if (tn + fp) > 0 else 0
    results_list.append({'Cutoff': round(cutoff, 2), 'Accuracy': acc, 'Sensitivity': sens, 'Specificity': spec})

cutoff_df = pd.DataFrame(results_list)
cutoff_df.head(10)
```

```python
# Plot accuracy, sensitivity, and specificity at different values of probability cutoffs
plt.figure(figsize=(10, 6))
plt.plot(cutoff_df['Cutoff'], cutoff_df['Accuracy'],    label='Accuracy',    linewidth=2)
plt.plot(cutoff_df['Cutoff'], cutoff_df['Sensitivity'], label='Sensitivity', linewidth=2)
plt.plot(cutoff_df['Cutoff'], cutoff_df['Specificity'], label='Specificity', linewidth=2)
plt.xlabel('Cutoff Value')
plt.ylabel('Score')
plt.title('Accuracy, Sensitivity and Specificity at Different Probability Cutoffs')
plt.legend()
plt.grid(True)
plt.show()

# Find optimal cutoff: where sensitivity and specificity are approximately equal
diff = abs(cutoff_df['Sensitivity'] - cutoff_df['Specificity'])
optimal_cutoff = cutoff_df.loc[diff.idxmin(), 'Cutoff']
print(f"Optimal cutoff (sensitivity ≈ specificity): {optimal_cutoff:.2f}")
```

#### **7.4.3** Get feature importance scores and select important features <font color="red">[2 Marks]</font>

```python
# Get feature importance scores from the trained model
feature_importances = rf_model.feature_importances_

# Create a DataFrame to visualise the importance scores
feat_imp_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feat_imp_df.head(20))
plt.title('Top 20 Feature Importances - Base Random Forest')
plt.tight_layout()
plt.show()

print(feat_imp_df.head(20).to_string(index=False))
```

```python
# Select features with high importance scores (importance >= 0.01)
importance_threshold = 0.01
important_features = feat_imp_df[feat_imp_df['Importance'] >= importance_threshold]['Feature'].tolist()
print(f"Number of important features (importance >= {importance_threshold}): {len(important_features)}")

# Create a new training data with only the selected features
X_train_rf = X_train[important_features].copy()
print(f"Training data shape with selected features: {X_train_rf.shape}")
```

#### **7.4.4** Train the model with selected features <font color="red">[1 Mark]</font>

```python
# Fit the model on the training data with selected features
rf_model_selected = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_model_selected.fit(X_train_rf, y_train_encoded)
print("Random Forest model trained on selected features successfully")
```

#### **7.4.5** Generate predictions on the training data <font color="red">[1 Mark]</font>

```python
# Generate predictions on training data
y_train_pred_rf = rf_model_selected.predict(X_train_rf)
```

#### **7.4.6** Check accuracy of the model <font color="red">[1 Mark]</font>

```python
# Check accuracy of the model
accuracy_rf_train = metrics.accuracy_score(y_train_encoded, y_train_pred_rf)
print("=" * 60)
print(f"Random Forest accuracy on training data: {accuracy_rf_train:.4f}")
print("=" * 60)
```

#### **7.4.7** Create confusion matrix <font color="red">[1 Mark]</font>

```python
# Create the confusion matrix to visualise the performance
conf_matrix_rf = metrics.confusion_matrix(y_train_encoded, y_train_pred_rf)
print("Confusion Matrix:")
print(conf_matrix_rf)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.title('Confusion Matrix - Random Forest (Training Data)')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

#### **7.4.8** Create variables for true positive, true negative, false positive and false negative <font color="red">[1 Mark]</font>

```python
# Create variables for true positive, true negative, false positive and false negative
TP_rf = conf_matrix_rf[1, 1]
TN_rf = conf_matrix_rf[0, 0]
FP_rf = conf_matrix_rf[0, 1]
FN_rf = conf_matrix_rf[1, 0]

print(f"True Positive  (TP): {TP_rf}")
print(f"True Negative  (TN): {TN_rf}")
print(f"False Positive (FP): {FP_rf}")
print(f"False Negative (FN): {FN_rf}")
```

#### **7.4.9** Calculate sensitivity, specificity, precision, recall and F1-score of the model <font color="red">[2 Marks]</font>

```python
# Calculate the sensitivity
sensitivity_rf = TP_rf / (TP_rf + FN_rf)
print(f"Sensitivity (Recall): {sensitivity_rf:.4f}")

# Calculate the specificity
specificity_rf = TN_rf / (TN_rf + FP_rf)
print(f"Specificity:          {specificity_rf:.4f}")

# Calculate Precision
precision_rf = TP_rf / (TP_rf + FP_rf)
print(f"Precision:            {precision_rf:.4f}")

# Calculate Recall
recall_rf = TP_rf / (TP_rf + FN_rf)
print(f"Recall:               {recall_rf:.4f}")

# Calculate F1 Score
f1_rf = 2 * precision_rf * recall_rf / (precision_rf + recall_rf)
print(f"F1 Score:             {f1_rf:.4f}")
```

#### **7.4.10** Check if the model is overfitting training data using cross validation <font color = "red">[2 marks]</font>

```python
# Use cross validation to check if the model is overfitting
cv_scores = cross_val_score(rf_model_selected, X_train_rf, y_train_encoded,
                            cv=5, scoring='accuracy', n_jobs=-1)
print("=" * 60)
print(f"Cross-validation accuracy scores: {cv_scores.round(4)}")
print(f"Mean CV accuracy: {cv_scores.mean():.4f}  (+/- {cv_scores.std():.4f})")
print(f"Training accuracy: {accuracy_rf_train:.4f}")
if accuracy_rf_train - cv_scores.mean() > 0.05:
    print("Warning: Model may be overfitting (training acc significantly > CV acc)")
else:
    print("Model generalises well - no significant overfitting detected")
print("=" * 60)
```
